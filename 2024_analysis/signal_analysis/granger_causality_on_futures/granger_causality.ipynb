# Plan
To evaluate whether Future prices have predictive power on ETF prices. For this, we use the Granger Causality Test (google it!).


Read the data

```python
import pandas as pd

order_book = pd.read_csv('[market_data_2] order_book.csv')

order_book['mid'] = order_book.apply(
    lambda row: (row['bidPrice0'] + row['askPrice0']) / 2,
    axis = 1
)
```

Run the granger causality test

```python
from statsmodels.tsa.stattools import grangercausalitytests

etf_mids = order_book[order_book['instrument'] == "ETF"][['mid', 'time']]
etf_mids.columns = ['etf', 'time']
future_mids = order_book[order_book['instrument'] == "Future"][['mid', 'time']]
future_mids.columns = ['future', 'time']

joined_mids = pd.merge(etf_mids, future_mids, on='time')

result = grangercausalitytests(joined_mids[['etf', 'future']], 1000)
```

# Results
The results are very clear. The Futures have an incredible strong predictive power on the ETF (p value of zero!!).
